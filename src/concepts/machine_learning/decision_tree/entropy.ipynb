```python
### Python Example

''' 

Python Code: 

''' 

from math import log2
from collections import Counter

def entropy_v1(labels: list[str]) -> float: # define entropy as a function. have labels that are
    # lists, will have some sort of floating point value at the end. 

    """
    Description: Calculating the entropy for a list of values.
    
    """

    # Entropy;  
    result_entropy = 0 #update zero as we go thru the loop. 

    # Get the proportions for the labels; (unique labels: 
    # count how many times they appear and divide by total number of labels to get proportion). 
    label_set = list(set(labels)) #set takes whatever collection you have as duplicates and doesn't return duplicates. 
    # can change it back to a list like above. 

    # Iterate through our unique labels
    for label in label_set:

        # Get the frequency of the label
        freq = labels.count(label)

        # Proportion
        p = freq / len(labels)

        # Product with the log transform
        ent = p * log2(p)

        # Adding the products together
        result_entropy = result_entropy + ent
    
    return -1 * result_entropy # takes -1 and muliplies it by the result_entropy to get the final entropy value.    


def entropy_v2(labels: list[str]) -> float:

    # Result for entropy
    result_entropy = 0

    # Counter (from collections framework to get the counts of each label). it counts how many times each unique label appears.
    counter = Counter(labels) #pass labels in counter. 

    # Get the label and associated count
    for count in counter.values():

        # Get the proportion
        p = count / len(labels) #len(labels) gives total number of labels.

        # Product with the log transform
        ent = p * log2(p)

        # Adding the products together
        result_entropy = result_entropy + ent
    
    return -1 * result_entropy


# Sample labels
samples = ['A', 'A', 'B', 'B', 'B', 'A']
sample_counter = Counter(samples)
print(sample_counter)

# Calculate entropy using version 1
print("Entropy v1:", entropy_v1(samples))

# Calculate entropy using version 2
print("Entropy v2:", entropy_v2(samples))
```
