```python
from functools import reduce

### Event Driven Functions: 

# Quickest way to get the numbers 1 through 10; 
nums = list(range(1, 11)) #make it 11 for 1-10. convert range to list. 


print(nums)

### Range and Lists; 
'''

range(1, 5)
This means:

1, 2, 3, 4


list() takes something iterable (like range) and turns it into a real list you can see.
list(range(1, 5))

Result:

[1, 2, 3, 4]


range() → generates numbers (lazy, efficient) / a recipe for numbers
list(range()) → actually creates a list of those numbers (stored in memory) / the finished dish with the actual numbers you can see and use

'''


zaed_nums_1 = range(1,10)
zaed_nums_2 = list(range(1,10))
print(zaed_nums_1)
print(zaed_nums_2)
```

```python
# Variable to store the result; 
result = 1


nums = list(range(1, 11)) #make it 11 for 1-10. convert range to list. 

print(nums)

# Iterate through the numbers; 
for num in nums: #for loop 

    # Square the num; 
    sq_num = num * num

    # Update the result; 
    result = result * sq_num # 1^2 * 2^2 * 3^2 ... *100^2. 


array_2 = 1*4*9*16*25*36*49*64*81*100

print(result)
print(array_2)
```

```python
# Event driven functions

from functools import reduce

"""
1. map()
2. reduce()
3. filter()

"""
### Done without a 'for' loop: 

## map()
sq_nums = map(lambda x: x**2, nums) # get all the squared numbers 

sq_nums_result = reduce(lambda a,b #input
                        : a*b #output
                        , map(lambda x: x**2, nums)) # get the product of all squared numbers.
print(f" Beta Test Result: {sq_nums_result}")




## Product
result = reduce(lambda a,b: a*b, map(lambda x: x**2, nums)) # multiply 
print(f"Product: {result}")

## Sum
result = reduce(lambda a,b: a+b, map(lambda x: x**2, nums)) # sum 
print(f"Sum: {result}")
```
